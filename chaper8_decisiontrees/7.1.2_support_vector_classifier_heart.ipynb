## Level 2: Fit model and print weights

```python
# Import packages and functions
import warnings
warnings.simplefilter("ignore")
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

# Load the heart dataset
heart = pd.read_csv("heart.csv")

# Create a dataframe X containing input features age and thalach
X = heart[["age", "thalach"]]

# Create a dataframe y containing output feature target
y = heart[["target"]]

# Create training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

# Scale the input features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize the model
SVCHeart = LinearSVC(C=1, penalty="l2", max_iter=3000, random_state=2)

# Fit the model (np.ravel reformats y_train as an array)
SVCHeart.fit(X_train, np.ravel(y_train))

# Print the model weights (intercept_ = w0, coef_ = w1 and w2)
print("w0:", SVCHeart.intercept_)
print("w1 and w2:", SVCHeart.coef_)
```

## Level 3: Confidence scores and accuracy

```python
# Import packages and functions
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

# Load the heart dataset
heart = pd.read_csv("heart.csv")

# Create a dataframe X containing input features trestbps and chol
X = heart[["trestbps", "chol"]]

# Create a dataframe y containing output feature target
y = heart[["target"]]

# Create training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

# Scale the input features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize the model
SVCModel = LinearSVC(C=5, penalty="l2", max_iter=3000, random_state=2)

# Fit the model
SVCModel.fit(X_train, np.ravel(y_train))

# Calculate and print confidence scores for testing set (instances with score < 1 are within margin)
confidence_scores = SVCModel.decision_function(X_test)
print(confidence_scores)

# Calculate and print proportion of instances correctly classified for testing set (.score() does this)
print(SVCModel.score(X_test, np.ravel(y_test)))
```
