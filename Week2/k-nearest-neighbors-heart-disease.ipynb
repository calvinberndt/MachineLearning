```python
# Import packages and functions
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Load the heart dataset
heart = pd.read_csv("heart.csv")

# Create a dataframe X containing input features thalach and age
X = heart[["thalach", "age"]]

# Scale the input features  
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a dataframe y containing output feature target
y = heart[["target"]]

# Initialize the model
heartModel = KNeighborsClassifier(n_neighbors=7)

# Fit the model
heartModel.fit(X_scaled, y)

# Make predictions
heartModel.predict(X_scaled)

# Calculate the accuracy of the model
print(heartModel.score(X_scaled, y))

# Print model parameters
print(heartModel.get_params())
```
